We’ll now build a neural network using the nn.Sequential approach in PyTorch. This method is easy and readable—perfect for beginners.

1: Import Required Libraries

```python
import torch
import torch.nn as nn
```

2: Define Network Parameters

```python
# Defining the input size, hidden layer size, output size, and batch size
n_in, n_h, n_out, batch_size = 10, 5, 1, 10
```

 3: Create Dummy Input and Target Tensors

```python
# Dummy input tensor of shape (10 samples × 10 features)
x = torch.randn(batch_size, n_in)

# Dummy target tensor ( 10 outputs) target is needed for supervised learning
y = torch.tensor([[1.0], [0.0], [0.0],
                  [1.0], [1.0], [1.0],
                  [0.0], [0.0], [1.0], [1.0]])
```

```python
x
```

4: create the Neural Network Model

```python
# Sequential model with 1 hidden layer and sigmoid activation at output
model = nn.Sequential(
    nn.Linear(n_in, n_h),  #This is both the "input interface" and the hidden layer 
    nn.ReLU(),
    nn.Linear(n_h, n_out),
    nn.Sigmoid()
)
```

This is a simple feedforward neural network with:

1 hidden layer <br>
ReLU activation in the hidden layer <br>
Sigmoid activation in the output layer



5: Define Loss Function and Optimizer

```python
# Mean Squared Error Loss
criterion = nn.MSELoss()

# Stochastic Gradient Descent optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

6: Train the Model with defined optimizer

```python
# Training loop
for epoch in range(50):
    # Forward pass
    y_pred = model(x)

    # Compute loss
    loss = criterion(y_pred, y)
    print('Epoch:', epoch, 'Loss:', loss.item())

    # Zero gradients before backward pass
    optimizer.zero_grad()

    # Backward pass (compute gradients)
    loss.backward()

    # Update model parameters
    optimizer.step()
```

## Cancer Prediction (melanoma Cancer Dataset)


### Load and Prepare the Data

```python
import os
import torch
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Check CUDA availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

```

```python
# Define data path
data_dir = "melanoma_cancer_dataset"  

# Load raw dataset (no transform yet)
raw_train_dataset = datasets.ImageFolder(os.path.join(data_dir, 'train'))
raw_test_dataset = datasets.ImageFolder(os.path.join(data_dir, 'test'))
```

### Visualize Sample Images

```python
from PIL import Image

def plot_images(dataset, class_names, num_images=8):
    fig, axes = plt.subplots(2, 4, figsize=(12, 6))
    axes = axes.flatten()

    for i in range(num_images):
        img, label = dataset[i]
        axes[i].imshow(img)
        axes[i].set_title(class_names[label])
        axes[i].axis('off')

    plt.suptitle("Training Images")
    plt.tight_layout()
    plt.show()

# Visualize from raw dataset
plot_images(raw_train_dataset, raw_train_dataset.classes)
```

### Data Augmentation/Transformation

```python
from torchvision import transforms

# Data Augmentation for Training
train_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.RandomHorizontalFlip(),
    # transforms.RandomRotation(15),
    # transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

# Simple Transform for Testing
test_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

# Assign transforms to existing raw datasets
raw_train_dataset.transform = train_transform
raw_test_dataset.transform = test_transform

# Create DataLoaders from updated datasets
train_loader = DataLoader(raw_train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(raw_test_dataset, batch_size=32, shuffle=False)
```

### Visualize Transformed (Augmented) Training Images

```python
import numpy as np

def plot_transformed_images(dataloader, class_names, num_images=8):
    # Get a batch from the dataloader
    images, labels = next(iter(dataloader))
    images = images[:num_images]
    labels = labels[:num_images]

    # Unnormalize for visualization
    images = images * 0.5 + 0.5  # Revert normalization

    fig, axes = plt.subplots(2, 4, figsize=(12, 6))
    axes = axes.flatten()

    for img, label, ax in zip(images, labels, axes):
        npimg = img.numpy().transpose((1, 2, 0))
        ax.imshow(np.clip(npimg, 0, 1))  # Ensure valid pixel range
        ax.set_title(class_names[label])
        ax.axis('off')

    plt.suptitle("Transformed & Augmented Training Images")
    plt.tight_layout()
    plt.show()

# Call the function using updated raw_train_dataset
plot_transformed_images(train_loader, raw_train_dataset.classes)
```

### Define Neural Network

```python
import torch.nn as nn
import torch.nn.functional as F

class SkinCancerNet(nn.Module):
    def __init__(self):
        super(SkinCancerNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.fc1 = nn.Linear(64 * 30 * 30, 128)
        self.fc2 = nn.Linear(128, 2)  # 2 classes: benign and malignant

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  # Conv1 -> ReLU -> Pool
        x = self.pool(F.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool
        x = x.view(-1, 64 * 30 * 30)          # Flatten
        x = F.relu(self.fc1(x))               # FC1 -> ReLU
        x = self.fc2(x)                       # Output layer
        return x

model = SkinCancerNet().to(device)
```

### Define Loss Function and Optimizer

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

### Training the Model

```python
!pip install tqdm
```

```python
from tqdm import tqdm  # Progress bar
import torch

num_epochs = 20
train_losses = []
train_accuracies = []

for epoch in range(num_epochs):
    running_loss = 0.0
    correct = 0
    total = 0

    print(f"\nEpoch {epoch+1}/{num_epochs}")
    for images, labels in tqdm(train_loader, desc=f"Training Epoch {epoch+1}", leave=False):
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        # Track accuracy
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    avg_loss = running_loss / len(train_loader)
    accuracy = 100 * correct / total
    train_losses.append(avg_loss)
    train_accuracies.append(accuracy)

    print(f"Loss: {avg_loss:.4f} | Train Accuracy: {accuracy:.2f}%")
```

### Evaluating the Model

```python
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Test Accuracy: {100 * correct / total:.2f}%")
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=raw_train_dataset.classes)
disp.plot(cmap='Blues')
plt.show()
```
